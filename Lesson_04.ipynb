### Median
The median is the middle value of the dataset


````r
median(data$wt)
````

### Mode

````r
# The mode is the number that is repeated most often in the dataset
get_mode <- function(x) {
   uniqv <- unique(x)
   uniqv[which.max(tabulate(match(x, uniqv)))]
}

get_mode(data$wt)
````

### First and third quartile
The first quartile shows the quarter or 25% point. Approximately 25% of the data values are less than or equal to the first quartile. The third quartile marks the upper 25% of data values.

````r
quantile(data$wt, 0.25)
quantile(data$wt, 0.75)
````

### Other quantiles
Why stop with just 1/4 of the data, we can work with deciles as well.

````r
quantile(data$wt, 0.4)
````

### Interquartile range
IQR is a difference between the first and the third quartile

````r
IQR(data$wt)
quantile(data$wt, 0.75) - quantile(data$wt, 0.25)
````

### Standard deviation and variance
The Standard Deviation is a measure of how spread out numbers are.
The variance is the average of the squared differences from the Mean.
Remember variance and standard deviation would be different whether you compute it for a population or a sample

![Image](https://upload.wikimedia.org/wikipedia/commons/thumb/8/8c/Standard_deviation_diagram.svg/2560px-Standard_deviation_diagram.svg.png)

````r
sd(data$wt)
var(data$wt)
````

## Basic visualizations of the data
- Boxplot
- Histogram
- QQ plot
- Scatter plot

````r
# Let's get another dataset
URL = 'https://raw.githubusercontent.com/lich-uct/r_tutorial/56e08976134ebbfaa7739e244306650841c204a9/06/alcohol_data.csv'
data <- read.csv(URL)
head(data, 3)
````

````r
# let's extract just the columns we will need
df_alcohol <- data[c('Country.Name', 'X2015..YR2015.')]
colnames(df_alcohol) <- c('Country', 'Consumation')
head(df_alcohol, 5)
````

````r
# Consumation column is a character type
df_alcohol$Consumation <- as.numeric(df_alcohol$Consumation)
````

````r
# we need to get rid of the missing values first and the change character column with Consumation numbers to a number type
df_alcohol <- df_alcohol[!is.na(df_alcohol$Consumation),]
df_alcohol$Consumation <- as.numeric(df_alcohol$Consumation)
````

````r
summary(df_alcohol)
````

````r
boxplot(df_alcohol$Consumation)
# Thick line is a median, bottom line is the first quartil and upper line is the third quartil
# Maximum and minimum are the lines coming from the box
````

````r
random_data <- c(rnorm(100, mean=10, sd=5), 28, 27, 26) # lets add some data to the plot

boxplot(df_alcohol$Consumation, random_data)
````

````r
# Dots shows the outliers - 1.5 times of the quartile range, this can be changed:
boxplot(df_alcohol$Consumation, random_data, range=3.0)
````

### Histogram

````r
Consumation <- df_alcohol$Consumation
hist(Consumation)
````

````r
hist(Consumation, breaks=30)
````

````r
df_alcohol[df_alcohol$Consumation >= 14, ]
````

````r
df_alcohol[df_alcohol$Consumation <= 1, ]
````

````r
df_consuments <- df_alcohol[df_alcohol$Consumation < 14 & Consumation > 1, ]
hist(df_consuments$Consumation, breaks = 15)
````

````r
# How would the histogram look if we had normal distribution of the data with the same parameters?
consuments_norm <- rnorm(
        nrow(df_consuments)
        , mean = mean(df_consuments$Consumation)
        , sd = sd(df_consuments$Consumation)
)
hist(consuments_norm, breaks = 15)
````

### QQ plot
 The Q-Q plot is a way for examination whether a set of data came from some theoretical distribution such as a Normal or not.

````r
qqnorm(df_consuments$Consumation, pch = 1, frame = FALSE)
qqline(df_consuments$Consumation, col = "steelblue", lwd = 2)
````

````r
qqnorm(consuments_norm, pch = 1, frame = FALSE)
qqline(consuments_norm, col = "steelblue", lwd = 2)
````

````r
install.packages('car')
library(car)

qqPlot(df_consuments$Consumation)
````

````r
ggplot(df_consuments, aes(sample=Consumation))+stat_qq()+theme_bw()
````

````r
scatterplot(mpg ~ wt | cyl, data=data,
   xlab="Weight of Car", ylab="Miles Per Gallon",
   main="Enhanced Scatter Plot",)
````

````r
# And the same advanced plot with ggplot
ggplot(data=mtcars, aes(x = wt, y = mpg, color = as.factor(cyl), shape = as.factor(cyl))) +
                          geom_point() + # plots the scatter plot
                          geom_smooth(method = "lm", se = F) + # plots the linear model
                          geom_smooth(se = F) + # plots the loess model
                          theme_minimal() # changes some of the formatting
````

### Line plot

````r
# Example data
x <- 1:10                                               
y <- rnorm(10, 5)

plot(x, y, type='l',                                 
     main='A Line Plot',
     xlab='X',
     ylab='Y',
     col='red', # line color
     lwd=8) # line thickness
````

````r
plot(x, y, type='b',                                 
     main='A Line Plot',
     xlab='X',
     ylab='Y',
     col='red', # line color
     lwd=8) # line thickness
````

````r
# create more data
y2 <- rnorm(10, 4)
y3 <- rnorm(10, 5)

# plot the first line
min_y <- min(c(y, y2, y3))
max_y <- max(c(y, y2, y3))
plot(x, y, type='b', col='orange', lwd=2, ylim=c(min_y, max_y))    

# add more lines
lines(x, y2, type='b', col='green', lwd=2, pch=10)
lines(x, y3, type='b', col='blue', lwd=2, pch=11) 


legend('topleft',                                      
       legend = c('Line y', 'Line y2', 'Line y3'),
       col = c('orange', 'green', 'blue'),
       pch = c(16, 10, 11),
       )
````

### PIE chart

````r
x <- c(1,1,2,3,2)
num <- c('first', 'second', 'third', 'fourth', 'fifth')
pie(x, labels=num, cex.lab=3)
````

````r
library('scales')
df_pie <- data.frame(x, num)
g <- ggplot(df_pie, aes(x='', y=num, fill=num)) +           # Leave x empty
      geom_bar(width=1, stat='identity', color='black',) +
      scale_fill_brewer(palette='Pastel2') +                # Custom colors
      coord_polar(theta='y', start=0) +                     # Polar coordinates
      theme(text=element_text(size=20), 
            axis.text = element_blank(),) +                 # make the text bigger and get rid of the axes tick labels
      geom_text(aes(x=1.4, label=num), 
                position=position_stack(vjust=0.4), size=5)
g

# save the plot
ggsave("pie_chart.png", height=4, width=6, dpi=150)
````
